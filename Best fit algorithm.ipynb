### Python implementation of the best fit algorithm using 

```python
import numpy as np; import pandas as pn; import matplotlib.pyplot as plt
# inputing of the Data
x = np.array([2,2.5,3,5,9])
y = np.array([-4,-5,2,1,24.3])
```

```python
space = np.linspace(2,9); w = 12; W = np.identity(5, dtype = float); n = len(x)
W[4], W[0] = w*W[4], w*W[0] 
A = np.column_stack((x,np.array([1,1,1,1,1])))
m = (n*(sum(x*y)) - sum(x)*sum(y))/(n*(sum(x**2)) - sum(x)**2)
b = (sum(x**2)*sum(y) - sum(x)*sum(x*y))/(n*sum(x**2)-sum(x)**2)
y_prediction = m*space + b
print('The best slope is ' + str(m) + ', and the best y-intersecpt is ' + str(b))
[m , b] = np.matmul(np.linalg.inv(np.matmul(np.matmul(np.transpose(A),W),A)),np.matmul(np.matmul(np.transpose(A),W),y))
print('The best fit weight slope is ' + str(m) + ', and the best weight y-intersecpt is ' + str(b))
y_weight_prediction = m*space + b
plt.figure(figsize= [7,5])
plt.plot(space, y_weight_prediction,'-.'),plt.plot(space, y_prediction), plt.plot(x,y,'go'), plt.xlabel('X values'), plt.ylabel('Y values')
plt.title('Y vs X with best fit lines'), plt.gca().legend(('Weighted','Normal'))
```
